```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

X = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 7],
    [7, 8],
    [8, 8]
])

y = np.array([0, 0, 0, 1, 1, 1])
```

```python
#ax + by + c =0
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()
learning_rate = 0.01
l = 1000
correct = 0
```

```python
def sigmoid(z):
    return 1/ (1+ np.exp(-z))
```

```python
for j in range(l):

    #p,q random point
    i = np.random.randint(0, len(X))
    p, q = X[i]
    yi = y[i]

    z = p*a +q*b + c
    y_hat = sigmoid(z)

    a+= learning_rate * p *(yi - y_hat)
    b+= learning_rate * q *(yi - y_hat)
    c+= learning_rate * (yi - y_hat)

    
```

```python
correct = 0
for i in range(len(X)):
    p, q = X[i]
    z = a*p + b*q + c
    y_hat = sigmoid(z)
    
    if y_hat >= 0.5:
        pred = 1
    
    else:
        pred = 0

    if pred == y[i]:
        correct += 1

print("Accuracy:", correct/len(X))
```
